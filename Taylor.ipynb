```python
var("x")
f1=log(x+1)
f2=exp(x)
def taylor_plot(f=f1, 
                x0=1, 
                k=3, 
                xintervall=(-1,4)):
    (xmin,xmax)=xintervall
    l = k+1
    p=plot(f,(x,xmin,xmax), 
           color=colors.darkred, 
           thickness=2,
           ymin=-2,
           ymax=3,)
    
    return(p+sum([plot(taylor(f,x,x0,n),
                       (x,xmin,xmax),
                       ymin=-1,
                       ymax=3,
                       color=colors.lightblue.blend(colors.darkblue,float(n)/l),
                       thickness=(2 if n==l-1 else 0.5)
                      ) for n in range(l)])
        +text(r"$T_{%d,%d}^f(x)=%s$"%(x0,k,str(latex(taylor(f,x,x0,k)))),
              (0,3.5),
              fontsize="medium",
              horizontal_alignment="left",
              color=colors.black)
        +text(r"$f(x)=%s$"%str(f),
              (0,3.9),
              horizontal_alignment="left",
              fontsize="medium",
              color=colors.black)
        )
    
def taylor_animation(f=f1, 
                     x0=1, 
                     k=slider(0,10,default=3),
                     xintervall=range_slider(-3,5,1,default=(-1,3))):
    show(taylor_plot(f=f,x0=x0,k=k, xintervall=xintervall))
interact(taylor_animation,f=f1)
```

```python
interact(taylor_animation,
         f=f2,
         x0=0,
         xintervall=range_slider(-4,4,1,default=(-2,2)))
```

```python
f1=exp(x)
for k in [0..6]:
    for x0 in [-2..2]:
        p=taylor_plot(f=f1,x0=1,k=k,xintervall=(-1,2))
        p.save_image("taylor-exp-%d-%d.png"%(x0,k))
```
